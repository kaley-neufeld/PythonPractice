```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
data_file = "./company_sales_data.csv"
df = pd.read_csv(data_file)
```

```python
df.head(5)
```

### 1. Read total profit of all months and show it using a line plot

```python
profitList = df['total_profit'].tolist()
monthList  = df['month_number'].tolist()
```

```python
plt.plot(monthList, profitList)
plt.xticks(monthList)
plt.yticks(range(0,550000,100000))
plt.xlabel('Month')
plt.ylabel('Profit (USD)')
plt.title('Company Profit per Month')
plt.show()
```

### 2. Plot total profit vs month with the following style properties:
* Line style is dashed and line-color should be red
* Show legend at the lower right location
* Add a circle marker
* Line marker color should be black
* Line width should be 3

```python
plt.plot(monthList, profitList,
         label = 'Profit Data over Time',
         color = 'r', marker = 'o',
         markerfacecolor = 'k',
         linestyle = '--',
         linewidth = 3)
plt.xticks(monthList)
plt.yticks(range(0,550000,100000))
plt.xlabel('Month')
plt.ylabel('Profit (USD)')
plt.title('Company Profit per Month')
plt.legend(loc='lower right')
plt.show()
```

### 3. Read all product sales data and show it  using a multiline plot
##### Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product )

```python
facecream_list = df['facecream'].tolist()
facewash_list = df['facewash'].tolist()
toothpaste_list = df['toothpaste'].tolist()
bathingsoap_list = df['bathingsoap'].tolist()
shampoo_list = df['shampoo'].tolist()
moisturizer_list = df['moisturizer'].tolist()
```

```python
plt.plot(monthList, facecream_list, label = 'Facecream Sales')
plt.plot(monthList, facewash_list, label = 'Facewash Sales')
plt.plot(monthList, toothpaste_list, label = 'Toothpaste Sales')
plt.plot(monthList, bathingsoap_list, label = 'BathingSoap Sales')
plt.plot(monthList, shampoo_list, label = 'Shampoo Sales')
plt.plot(monthList, moisturizer_list, label = 'Moisturizer Sales')

plt.xticks(monthList)
plt.yticks(range(0,15000,1000))
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.title('Sales per Product per Month')
plt.legend(loc='upper left')
plt.show()

# todo: facewash show on plot?
```

### 4. Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(monthList, toothpaste_list)
plt.xticks(monthList)
plt.xlabel('Month')
plt.ylabel('Number of Toothpaste Units Sold')
plt.title('Toothpaste Sales over Time')
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

### 5. Read face cream and facewash product sales data and show it using the bar chart

```python
plt.bar([a - 0.25 for a in monthList], facecream_list, width= 0.25, label = 'Face Cream', align='edge')
plt.bar([a + 0.25 for a in monthList], facewash_list, width= -0.25, label = 'Face Wash', align='edge')
plt.xticks(monthList)

plt.xlabel('Month')
plt.ylabel('Number of Product Sold')
plt.legend(loc='upper left')
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash vs Facecream Sales Data')

plt.show()
```

### 6. Read sales data of bathing soap of all months and show it using a bar chart. Save this plot

```python
plt.bar(monthList, bathingsoap_list)
plt.xticks(monthList)

plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.grid(True, linewidth=1, linestyle="--")
plt.title('Bathing Soap Sales Data')

# un-comment below to save figure!
# plt.savefig('./bathingsoap_sales_data.png')
plt.show()
```

### 7. Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
max_profit = max(profitList)
min_profit = min(profitList)
num_bins = 4
jumps = (max_profit - min_profit)//num_bins

plt.hist(profitList, bins=num_bins)
plt.yticks(range(0,10,2))
plt.xticks(range(min_profit,max_profit+10000,jumps))
plt.xlabel('Profit Ranges (USD)')
plt.ylabel('Number of Months')
plt.title('Popularity of Profit Ranges')
plt.show()
```
